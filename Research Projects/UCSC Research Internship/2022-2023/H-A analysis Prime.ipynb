```python
import csv
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data4 = pd.read_csv('Downloads/H-Alpha data sheet - Data Table.csv',sep = ',',header = 0)
Data5 = pd.read_csv('Downloads/Smith and Dupree, Table 1 Data Values.csv',sep = ',', names = ['names','Plx','Gmag','BPmag','RPmag','Jmag','Kmag'])
```

```python
distance = []
G_color = []
M_G = []
M_RP = []
M_BP = []
JK_color = []
M_K = []

for item in Data5['Plx']:
    distance.append(1/(abs(float(item))/1000))
for i in range(len(distance)):
    G_color.append(float(Data5['BPmag'][i])-float(Data5['RPmag'][i]))
    M_G.append(float(Data5['Gmag'][i])-5*math.log(distance[i],10)+5)
    M_RP.append(float(Data5['RPmag'][i])-5*math.log(distance[i],10)+5)
    M_BP.append(float(Data5['BPmag'][i])-5*math.log(distance[i],10)+5)
    JK_color.append(float(Data5['Jmag'][i])-float(Data5['Kmag'][i]))
    M_K.append(float(Data5['Kmag'][i])-5*math.log(distance[i],10)+5)

Data5['distance'] = distance
Data5['G_color'] = G_color
Data5['M_G'] = M_G
Data5['M_RP'] = M_RP
Data5['M_K'] = M_K
Data5['JK_color'] = JK_color
```
